```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import statsmodels.api as sm
from sklearn.impute import SimpleImputer  # Import SimpleImputer

# Assuming X is the feature matrix and y is the target vector
model = LinearRegression()
model.fit(X_train, Y_train)

# Make predictions
y_pred = model.predict(X_train)

# Calculate R^2, MAE, MSE, RMSE
r2 = r2_score(Y_train, y_pred)
mae = mean_absolute_error(Y_train, y_pred)
mse = mean_squared_error(Y_train, y_pred)
rmse = np.sqrt(mse)

# Handle missing or infinite values in X before adding a constant
X_imputed = X.copy()  # Create a copy of X to avoid modifying the original DataFrame

# Replace infinite values with NaN
X_imputed.replace([np.inf, -np.inf], np.nan, inplace=True)

# Impute NaN values using the mean
imputer = SimpleImputer(strategy='mean')
X_imputed = pd.DataFrame(imputer.fit_transform(X_imputed), columns=X.columns)

# For more detailed statistics (including p-values, VIF)
X_with_const = sm.add_constant(X_imputed)  # Add constant to the imputed data
sm_model = sm.OLS(Y_train, X_with_const.loc[Y_train.index])  # Use .loc to select rows corresponding to Y_train
results = sm_model.fit()

# Print model statistics
print(f"R^2: {r2}")
print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"Adjusted R^2: {results.rsquared_adj}")
print(results.summary())  # This gives p-values, F-statistic, etc.
```
